```python
#Answert to Part 1

#Our Payoff Matrix M m=[[R S],[T P]] T=3,R=2,P=1,s=0 which satisfy our conditions T>R>P>S , 2R>T+S
payoff_Matrix=np.matrix([[2 , 0],[3 , 1]]) #Define Payoff matrix 

#Function to implement Nice Guy Strategy which always cooperate. 
#This function just will generate array of 2 values[1,0] 
#This function will return array [1,0] when we call it.
def nice_guy():
    ng = np.array([1,0])
    return ng

#Function to implement Bad Guy Strategy which always defect. 
#This function just will generate array of 2 values[0,1] 
#This function will return array [0,1] when we call it.
def bad_guy():
    bg = np.array([0,1])
    return bg

#Function to implement Mainly Nice Guy Strategy which cooperate and defect  randomly. 
#This function will generate array of 2 values[0,1] or [1,0] which can choose randomly
#This function will generate array of 2 values => 30% will be [0,1] and 70% will be  [1,0].
#This function will return array 30% will be [0,1] and 70% will be  [1,0] when we call it.
def mainly_nice():
    #we use random.choice to create binary array with size 2 with probability 0=>30% , 1=>70% 
    #random.choice(array of [0 1], size of array(shape), p=prob for first and second values in array)
    mn = np.random.choice(np.arange(2), 2 ,0, p=[0.3, 0.7]) 
    return mn

#Function to implement Mainly Bad Guy Strategy which cooperate and defect  randomly. 
#This function will generate array of 2 values[0,1] or [1,0] which can choose randomly
#This function will generate array of 2 values => 70% will be [0,1] and 30% will be  [1,0].
#This function will return array 70% will be [0,1] and 30% will be  [1,0] when we call it.
def mainly_bad():
    #we use random.choice to creat binary array with size 2 with probability 0=>70% , 1=>30% 
    #random.choice(array of [0 1], size of array(shape), p=prob for first and second values in array)
    mb = np.random.choice(np.arange(2), 2 ,0, p=[0.7, 0.3]) 
    return mb

#Function to implement Tit for Tat Strategy which start by cooperating, then repeat what the opponent has done in the previous move. 
#This function will generate array of 2 values[0,1] or [1,0] depend on opponent move.
#This function take 2 parameters,
    #parameter 1 which is represent the round number is game.
    #parameter 2 which is opponent strategy to be repeated.
#At the first round this function will generate nice guy strategy, its mean generate array of [1,0]
#after rgw first round it will take opponent strategy and will repeate it , it mean will repeate opponent array
#This function will return array that depend in round number and opponent strategy.
def tit_for_tat(round_num=1,opponent_strategy=nice_guy):
    if round_num==1:
      tft = np.array([1,0])
    else:
      tft = opponent_strategy    
    return tft 

#This function to calculate the reward in every round, and will take 2 values
    #u1 which is first strategy (array of [1,0] or [0,1])
    #u2 which is second strategy (array of [1,0] or [0,1])
#This function will calculate the payoff for player1 and payoff for player2
#The mechanism of calculation for 2 players as the following 
    #Take the first array and transpose it and multiply it by payoff matrix and second array (u1T*payoff_Matrix*u2)
    #Take the second array and transpose it and multiply it by payoff matrix and first array (u2T*payoff_Matrix*u1)
#This function will return 2 values(the payoff for first player and the payoff of second player)
def reward(u1,u2):
    
    #r1=u1T*payoff_Matrix*u2
    #u1=u1.transpose()
    r1=0 #inital value for reward for player 1
    r2=0 #inital value for reward for player 2
    r1=np.dot(np.dot(u1,payoff_Matrix),u2)
    #r2=u2T*payoff_Matrix*u1
    # u2=u2.transpose()
    r2=np.dot(np.dot(u2,payoff_Matrix),u1)
    return r1,r2

#simple example to implement 2 strategies Nice guy , Nice guy for 3 Iterations

cumulative_r1=0 #Initial value
cumulative_r2=0 #Initial value

#The Game will start here
#Number of rounds is 3

print("Using Nice Guy Strategy for both players\n")

for i in range(1,4):

    r1,r2=reward(nice_guy(),nice_guy())
    cumulative_r1=cumulative_r1+r1
    cumulative_r2=cumulative_r2+r2
    print("Round Number "+str(i )+str(r1)+str(r2))

#Some Conclusions
print("\nThe Last Reselt is "+str(cumulative_r1)+str(cumulative_r2)) 
```

```python
#Answert to Part 4

#Player 2 will be genetic mutation and will have random gene 
#Here we have scenario of the first player fixed his strategy (nice guy) for 20 rounds
#The oponent has gene value and this value will be generated automatically using random variable, 
#In this game we will found the type of opponent during his playing in 20 rounds 
#Player 1 will measure the collaboration for player 2 with fixed strategy "Nice Guy"

#Here we generate random Gene to measure the player's cooperation
gene=round(random.uniform(0, 1),1)

#We generate array to make history for our game (to save what will happen in every round)
play_history=[] #Initial array
cumulative_r1=0 #Initial value
cumulative_r2=0 #Initial value
diff_Payoff=0 #Initial value

#We make function called player_type, this function take array and estimate the type of opponent strategy 
#the type is 
# 1-nice_one
# 2-bad_one
# 3-mainly_bad
# 4-mainly_nice

def player_type(play_history):

  nice_count=0
  bad_count=0
  for i in range(len(play_history)):        
    if play_history[i][0]==0:
      bad_count=bad_count+1
    elif play_history[i][0]==1:
      bad_count=bad_count+1
    elif play_history[i][0]==2:
      nice_count=nice_count+1
    elif play_history[i][0]==3:
      nice_count=nice_count+1

  if bad_count==0:
    player_ty="nice_one"
  elif nice_count==0:
    player_ty="bad_one"
  elif bad_count>len(play_history)/2 and bad_count < len(play_history):
    player_ty="mainly_bad"
  elif nice_count >= len(play_history)/2 and nice_count < len(play_history) :
    player_ty="mainly_nice"

  return player_ty

#Here we define a new function, this function take the gene value and estimate the type of strategy will be used for that player 
#here we have 4 scenarios
# 1- gene between 0 and .25 => its mean use bad_guy stratergy.
# 2- gene between .25 and .5 => its mean use meainly_bad stratergy.
# 3- gene between .5 and .75 => its mean use meainly_nice stratergy.
# 4- gene between .75 and 1 => its mean use nice_guy stratergy.

def gene_strategy(gene):

  if gene>=0 and gene<.25:
    strategy1=bad_guy()
  elif gene>=.25 and gene<.5:
    strategy1=mainly_bad()
  elif gene>=.5 and gene<.75:
    strategy1=mainly_nice()
  elif gene>=.75 and gene<=1:
    strategy1=nice_guy()

  return strategy1
  
#Here we define new function that take 2 values which are the total payoff for player 1 and total payoff player 2 
#and calculate the total payoff for whole game and the difference in payoff between the 2 players for whole game

def final_Result(cumulative_r1,cumulative_r2):

  total_Payoff=cumulative_r1+cumulative_r2
  diff_Payoff=cumulative_r1-cumulative_r2
  print("Total Payoff for Player1 is "+str(cumulative_r1))
  print("Total Payoff for Player2 is "+str(cumulative_r2))
  print("Total Payoff for both Players is "+str(total_Payoff))

  if cumulative_r1>cumulative_r2:
    print("The Winner is Player1: "+str(cumulative_r1)+" (The Differance is "+str(abs(diff_Payoff))+")")
  if cumulative_r1<cumulative_r2:
    print("The Winner is Player2: "+str(cumulative_r2)+" (The Differance is "+str(abs(diff_Payoff))+")")
  if cumulative_r1==cumulative_r2:
    print("Equal Result for Both Player1 and Player2 "+str(cumulative_r1)+" (The Differance is "+str(abs(diff_Payoff))+")")

  return diff_Payoff

#The Game will start here
#Number of rounds is 20 
#First player always play nice guy strategy for 20 rounds
#The second player will use gene value to choose his strategy which will play for 20 rounds with this strategy
#m is the number of rounds m=20
#r1 is payoff of player 1
#r2 is payoff of player 2

for m in range(1,21):
  
  r1,r2=reward(nice_guy(),gene_strategy(gene)) 
  play_history.append([int(r1),int(r2)])

#To count total payoff for player 1 and for player 2 after the game ended

for i in range(len(play_history)):
  cumulative_r1=cumulative_r1+play_history[i][0]
  cumulative_r2=cumulative_r2+play_history[i][1]

#Some Conclusions
print("Gene for player1 is: "+str(gene))
print("playing History for our game for 20 Rounds is: ")
print(play_history)    
print("\nAfter playing 20 rounds, the predicting strategy for Opponent player (player2) is: "+ str(player_type(play_history))+"\n")

#call function final_Result
diff_Payoff_game1=final_Result(cumulative_r1,cumulative_r2)
```
